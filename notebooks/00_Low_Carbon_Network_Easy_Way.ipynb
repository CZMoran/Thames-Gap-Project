```python
## Set up relative paths for data and output
## Run this cell each time you start a new session to ensure paths are correctly set

from pathlib import Path

## Set up base path for the project
BASE_DIR = Path("/Users/zc/Desktop/Projects/Project_1/notebooks/00_Low_Carbon_Network_Easy_Way.ipynb").resolve().parent.parent

## Define paths for raw data, processed data, and outputs
raw_data = BASE_DIR / "data_raw"
processed_data = BASE_DIR / "data_processed"

## Define specific file paths
OSM = raw_data / "OSM"
```

```python
# Only run it once to download the data, then comment it out to avoid repeated downloads
# Walking network
London_walk = ox.graph_from_place(
    "London, England",
    network_type='walk',
    retain_all=False  # keeps only largest connected component automatically
)

# Cycling network
London_cycle = ox.graph_from_place(
    "London, England",
    network_type='bike',
    retain_all=False
)

# Car network
London_car = ox.graph_from_place(
    "London, England",
    network_type='drive',
    retain_all=False
)

# Save as GeoPackage
ox.save_graph_geopackage(London_walk, filepath='OSM/walk_network.gpkg')
ox.save_graph_geopackage(London_cycle, filepath='OSM/cycle_network.gpkg')
ox.save_graph_geopackage(London_car, filepath='OSM/car_network.gpkg')
```
